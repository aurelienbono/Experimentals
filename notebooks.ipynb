```python
import random
from datetime import datetime, timedelta
from uuid import uuid4

# Fonction pour générer un faux URL d'image
def generate_fake_image_url(image_number):
    return f"https://example.com/images/fake_baggage_{image_number}.jpg"

# Générer des données fictives pour 5 réservations de transport
fake_rides_reservations = []

for _ in range(5):  # Générer 5 ensembles de données fictives
    logistic_id = str(uuid4())  # Générer un identifiant unique
    logistic_type = "Transport - Taxi"
    status = random.choice(["En cours", "Livré", "Annulé"])
    created_at = datetime.now() - timedelta(days=random.randint(1, 30))
    pickup_location = f"Lieu de ramassage {random.randint(1, 10)}"
    dropoff_location = f"Lieu de dépôt {random.randint(1, 10)}"
    distance = random.randint(3, 10)
    seats = random.choice(["Lourd", "lege", "moyen"])
    description_of_luggage = random.choice(["Petit sac", "Valise moyenne", "Grand coffre"])
    mode_of_payments = random.choice(["Espèces", "Carte bancaire", "Mobile Money"])
    client_id = str(uuid4())  # Simuler un ID client unique

    # Générer des URLs fictifs pour les images de bagages
    baggage_image_1 = generate_fake_image_url(1)
    baggage_image_2 = generate_fake_image_url(2)
    baggage_image_3 = generate_fake_image_url(3)

    rides_reservations_data = {
        "id": logistic_id,
        "type": logistic_type,
        "status": status,
        "created_at": created_at.strftime("%Y-%m-%d %H:%M:%S"),
        "pickup": pickup_location,
        "baggage_image_1": baggage_image_1,
        "baggage_image_2": baggage_image_2,
        "baggage_image_3": baggage_image_3,
        "clients": client_id,
        "package_weight": seats,
        "distance": distance,
        "description_of_luggage": description_of_luggage,
        "mode_of_payments": mode_of_payments,
        "dropout": dropoff_location,  # Correction de "dropout" -> "dropoff"
        "estimation_time": str(timedelta(hours=random.randint(1, 5))),  # Convertir en string
        "trajet": f"{pickup_location} - {dropoff_location}",
    }

    fake_rides_reservations.append(rides_reservations_data)

# Simuler une réponse API
api_response = {
    "message": "Taxi reservation details retrieved successfully.",
    "data": fake_rides_reservations
}

# Affichage du résultat au format JSON
import json
print(json.dumps(api_response, indent=4, ensure_ascii=False))
```
